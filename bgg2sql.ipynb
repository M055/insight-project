## Create 2D maytrx

```python
finalgamelist_df.loc[finalgamelist_df['idx']==10000,:]
```

```python
idx = []
row_ids = []
col_ids = []
sim_score = []
gamename=[]

ii = 1
for dumrowid in finalgamelist_df['idx']:
    for dumcolid in finalgamelist_df['idx']: # Both refer to the same game..
        if dumrowid >= dumcolid:
            curr_gamename1 =  finalgamelist_df.loc[dumrowid,'gamename']
            curr_gamename2 =  finalgamelist_df.loc[dumcolid,'gamename']
            curr_simscore = allgamesimilarities_df.loc[curr_gamename1,curr_gamename2]
            
            idx.append(ii)
            row_ids.append(dumrowid)
            col_ids.append(dumcolid)
            sim_score.append(curr_simscore)
            gamename.append(curr_gamename1)
            
            ii+=1
```
